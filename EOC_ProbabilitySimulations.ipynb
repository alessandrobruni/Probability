
**1. make 100000 simulations to:
12 boxes, 10 balls put in boxes at random. what is the probability to find 5 boxes empty?**

```python
def boxes(number_of_boxes = 12, number_of_balls = 10, target=5):
    """Make NUM_SIMULATIONS (100_000) simulations to:
    Having number_of_boxes boxes, and putting boxes at random number_of_balls balls,
    what is the probability to find target boxes empty?

    :param number_of_boxes: default 12
    :param number_of_balls: default 19
    :param target: default 5
    :return:  the probability to find target boxes empty
    """
    bask =np.zeros(number_of_boxes,int) #prepare the boxes into which put the balls
    count_zeros = 0

    for loop in range(NUM_SIMULATIONS):

        for l in range(number_of_balls):# launch the 10 times
            bask[np.random.randint(0,number_of_boxes)] += 1 # ball adding them to the basket
        #print(bask)

        if bask[ bask == 0].size == target : # if num of empty basket equals the target, remember it
            count_zeros += 1

        bask =np.zeros(12,int)  #new empty boxes

        #print(f"loop {loop} - {count_5zero}")
    p_zero= count_zeros/NUM_SIMULATIONS

    print(f"Having {number_of_boxes} boxes and putting inside randomly {number_of_balls} of balls, leads to a probability of {p_zero} to find {target} box empty. ")

    return p_zero

print(boxes())
```

**2. make 100000 simulations to:
The game consists of N rounds,
where P is a probability of winning in a single round.
Each game stops when the player lose in 2 rounds (not necessarily sequentially).
What is the average duration of a game (in rounds)?
Run a simulation for  P=0.6**


```python
def game_2_loses( p = 0.6, rounds_to_lose = 2):
    """Make NUM_SIMULATIONS (100_000) simulations to:
    The game consists of N rounds, where P is a probability of winning in a single round.
    Each game stops when the player lose in 2 rounds (not necessarily sequentially).
    What is the average duration of a game (in rounds)?

    :param p: the probability to lose , default 0.6
    :param rounds_to_lose: stop the game when player lose rounds_to_lose  , default 2
    :return:
    """
    runs = []
    for round in range(NUM_SIMULATIONS):
        runned = 0 #take counts of the played round until 2 lose happens
        lose = 0   #take counts of the loosing games
        play = True
        while  play:
            game = np.random.binomial(1,p) #do the game
            lose +=1 if game == 0 else 0   # if losed , remember it

            play = False if lose == rounds_to_lose else True #continue to play until rounds to lose
            runned += 1

        runs.append(runned) #remember the rounds played

    pbr = np.array(runs).mean()

    print(f"The {p} propability of wining a single round and stop the game after {rounds_to_lose} loses leads to a average duration of the game equal to {pbr}. ")

    return pbr

print(game_2_loses())
```

**3. you have a deck of 100 cards with values from 1 to 100.
you draw 2 cards without replacement. 
what is probability that the number on one card is double than the other?
(this is an interview question from Zenfit)**

```python
def cards(number_of_cards = 100 , cards_drawn = 2 ):
    """You have a deck of 100 cards with values from 1 to 100.
    You draw 2 cards without replacement.
    What is probability that the number on one card is double than the other?

    :param number_of_cards:
    :param cards_drawn: fixed to 2
    :return:
    """
    cards = np.array(range(1,number_of_cards+1)) # the deck

    doubles=[]  #  the events of doubling

    for j in range(NUM_SIMULATIONS):
        game = np.sort(np.random.choice(cards,size=2,replace=False)) #chose  2 cards without replacement
        doubles.append(game[1]/game[0] == cards_drawn) # the cards are sorted so just divide to see if double

    prb_doubling = np.array(doubles).mean()

    print(f"The propability to draw 2 cards from a deck of {number_of_cards} cards with values from 1 to {number_of_cards}  and that the number on one card is double than the is {prb_doubling}. ")


    return prb_doubling

print(cards())
```
